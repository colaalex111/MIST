```python
import numpy as np
np.set_printoptions(suppress=True)

data = np.load('./expdata/cifar10_resnet18_2_100_0_0_0_0_0.0_1_0.0_logit_diff_0.0_25.0_1_loss_distribution.npy')
label = np.load('./expdata/cifar10_resnet18_2_100_0_0_0_0_0.0_1_0.0_logit_diff_0.0_25.0_1_loss_label.npy')


#data = np.load('./expdata/cifar10_alexnet_5_40_0_0_0_0_0.0_1_0.1_logit_diff_0.0_metric_distribution.npy')
#label = np.load('./expdata/cifar10_alexnet_5_40_0_0_0_0_0.0_1_0.1_logit_diff_0.0_metric_label.npy')
```

```python
print (data.shape,label.shape)
all_metric_result = data
#all_metric_result = np.log(data*-1)
```

```python
all_labels = label
all_metric_result = data
fpr_threshold = 0.01
from sklearn.metrics import roc_auc_score
pos_index = np.arange(len(all_labels))[all_labels == 1]
neg_index = np.arange(len(all_labels))[all_labels == 0]
print (len(pos_index),len(neg_index))
min_len = min(len(pos_index),len(neg_index))
sampled_pos_index = np.random.choice(pos_index,min_len,replace=False)
sampled_neg_index = np.random.choice(neg_index,min_len,replace=False)
sampled_index = np.concatenate((sampled_neg_index,sampled_pos_index),axis=0)
all_metric_result = all_metric_result[sampled_index]
all_labels = all_labels[sampled_index]
#print (all_metric_result.shape,all_labels.shape)
#print (all_metric_result[:100],all_labels[:100])
## calcualte AUC and PLR
auc = roc_auc_score(all_labels,all_metric_result)
print(f"AUC score {auc}")
negative_index = np.arange(len(all_labels))[all_labels == 0]
threshold_index = int(len(negative_index)*fpr_threshold)
threshold = np.sort(all_metric_result[negative_index])[::-1][threshold_index]
print(f"Likelihood threshold {threshold}")

print (f"threshold index {threshold_index}")
cnt = 0
for i in range(len(all_metric_result)):
    if (all_metric_result[i] >= threshold and all_labels[i]==1):
        cnt += 1
print(f"TPR {cnt/len(negative_index)}, FPR {fpr_threshold}, PLR {cnt / threshold_index}")

```

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(all_labels, all_metric_result, pos_label=1)
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(5,5))
plt.plot(fpr,tpr)
plt.plot(fpr,fpr)
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.legend(['roc curve','y=x baseline'])
plt.show()
```

```python
min_val = np.amin(all_metric_result)
max_val = np.amax(all_metric_result)

print (min_val,max_val)
```

```python

ranges = np.linspace(-2,0,num=200)

pos_metric = all_metric_result[np.arange(len(all_labels))[all_labels == 1]]
neg_metric = all_metric_result[np.arange(len(all_labels))[all_labels == 0]]
pos_count,_ = np.histogram(pos_metric,bins=ranges)
neg_count,_ = np.histogram(neg_metric,bins=ranges)
#print (pos_count)
#print (neg_count)
```
