```python

def all_analysis(epochs,name_prefix,dataset,model,eval_data_size,special_layers=None,num_layers=12,num_user=5):
    
    total_data_num = (num_user+1)*eval_data_size
    all_epoch_cos = np.zeros((total_data_num,len(epochs),num_user))
    all_epoch_loss = np.zeros((total_data_num,len(epochs),num_user))
    all_epoch_target_loss = np.zeros((total_data_num,len(epochs),num_user))
    all_epoch_grad_diff = np.zeros((total_data_num,len(epochs),num_user))
    all_epoch_layer_cos = np.zeros((total_data_num,len(epochs),num_user,num_layers))
    all_epoch_layer_grad_diff = np.zeros((total_data_num,len(epochs),num_user,num_layers))
    all_epoch_target_after_loss = np.zeros((total_data_num,len(epochs),num_user))
    all_epoch_label = np.zeros(total_data_num)
    best_layer_dict = {'cifar100':6,'cifar10':2}
    this_best_layer = best_layer_dict[dataset]
    
    for epoch_idx,epoch in enumerate(epochs):
        data_name = prefix+'all_info_multi_party_member_attack_'+str(num_user)+'_0_0.0_0_0.0_0_'+str(epoch+1)+'_'+str(dataset)+'_'+str(eval_data_size)+'_'+str(eval_data_size)+'_'+str(model)+'.npy'
        data = np.load(data_name,allow_pickle=True)
        label_name = prefix+'all_label_multi_party_member_attack_'+str(num_user)+'_0_0.0_0_0.0_0_'+str(epoch+1)+'_'+str(dataset)+'_'+str(eval_data_size)+'_'+str(eval_data_size)+'_'+str(model)+'.npy'
        label = np.load(label_name)
        loss_data_name = prefix+'loss_info_multi_party_member_attack_'+str(num_user)+'_0_0.0_0_0.0_0_'+str(epoch+1)+'_'+str(dataset)+'_'+str(eval_data_size)+'_'+str(eval_data_size)+'_'+str(model)+'.npy'
        loss_data = np.load(loss_data_name,allow_pickle=True)
        loss_label_name = prefix+'loss_label_multi_party_member_attack_'+str(num_user)+'_0_0.0_0_0.0_0_'+str(epoch+1)+'_'+str(dataset)+'_'+str(eval_data_size)+'_'+str(eval_data_size)+'_'+str(model)+'.npy'
        loss_label = np.load(loss_label_name,allow_pickle=True)
        
        target_loss_data_name = prefix+'target_model_before_loss_info_multi_party_member_attack_'+str(num_user)+'_0_0.0_0_0.0_0_'+str(epoch+1)+'_'+str(dataset)+'_'+str(eval_data_size)+'_'+str(eval_data_size)+'_'+str(model)+'.npy'
        target_loss_data = np.load(target_loss_data_name,allow_pickle=True)
        target_after_loss_data_name = prefix+'target_model_after_loss_info_multi_party_member_attack_'+str(num_user)+'_0_0.0_0_0.0_0_'+str(epoch+1)+'_'+str(dataset)+'_'+str(eval_data_size)+'_'+str(eval_data_size)+'_'+str(model)+'.npy'
        target_after_loss_data = np.load(target_loss_data_name,allow_pickle=True)
        data = np.reshape(data,(-1,num_user,12,14))
        label = np.reshape(label,(-1))
        loss_data = np.reshape(loss_data,(-1,num_user))
        loss_label = np.reshape(loss_label,(-1))
        target_loss_data = np.reshape(target_loss_data,(-1,1))
        target_loss_data = np.tile(target_loss_data,(1,num_user))
        target_after_loss_data = np.reshape(target_after_loss_data,(-1,1))
        target_after_loss_data = np.tile(target_after_loss_data,(1,num_user))
        #print (data.shape,label.shape,loss_data.shape,loss_label.shape,target_loss_data.shape)
        
        all_epoch_cos[:,epoch_idx,:] = data[:,:,this_best_layer,0]
        all_epoch_grad_diff[:,epoch_idx,:] = data[:,:,this_best_layer,1] - data[:,:,this_best_layer,2]
        all_epoch_layer_cos[:,epoch_idx,:,:] = data[:,:,:,0]
        all_epoch_layer_grad_diff[:,epoch_idx,:,:] = data[:,:,:,1] - data[:,:,:,2]
        all_epoch_loss[:,epoch_idx,:] = loss_data
        all_epoch_target_loss[:,epoch_idx,:] = target_loss_data
        all_epoch_target_after_loss[:,epoch_idx,:] = target_after_loss_data
        all_epoch_label = label
        
    return all_epoch_layer_cos
```

```python

prefix = '/Users/jclialex/PycharmProjects/whiteboxmi_expdata/expdata/expdata/'
epochs = ((np.arange(29)+1)*10)
data_a = all_analysis(epochs,prefix,'cifar10','alexnet',5000,num_user=3,special_layers=True)
fig = plt.figure(figsize=(24,12))
print (data_a.shape)

data = data_a[:5000,:,0,:]
plt.subplot(1,3,1)
for layer in range(12):
    if (layer%2 == 0):
        plt.plot(epochs,np.average(data[:,:,layer],axis=0))
    
plt.legend([0,2,4,6,8,10], prop={'size': 18})
plt.title('Alexnet-Cifar-10',fontsize=18)
plt.xlabel('epochs', fontsize=18)
plt.ylabel('mean of cosine similarity', fontsize=18)
#plt.show()

plt.subplot(1,3,2)
for layer in range(12):
    if (layer%2 == 0):
        plt.plot(epochs,np.var(data[:,:,layer],axis=0))
    
plt.legend([0,2,4,6,8,10], prop={'size': 18})
plt.title('Alexnet-Cifar-10',fontsize=18)
plt.xlabel('epochs', fontsize=18)
plt.ylabel('var of cosine similarity', fontsize=18)

plt.subplot(1,3,3)
for layer in range(12):
    if (layer%2 == 0):
        plt.plot(epochs,np.average(data[:,:,layer],axis=0)/np.var(data[:,:,layer],axis=0))
    
plt.legend([0,2,4,6,8,10], prop={'size': 18})
plt.title('Alexnet-Cifar-10',fontsize=18)
plt.xlabel('epochs', fontsize=18)
plt.ylabel('mean/var of cosine similarity', fontsize=18)

plt_name = 'cos_vis_alexnet_per_layer_member.pdf'
plt.savefig(plt_name,dpi=800, bbox_inches="tight")
plt.show()

fig = plt.figure(figsize=(24,12))
data = data_a[-5000:,:,0,:]
plt.subplot(1,3,1)
for layer in range(12):
    if (layer%2 == 0):
        plt.plot(epochs,np.average(data[:,:,layer],axis=0))
    
plt.legend([0,2,4,6,8,10], prop={'size': 18})
plt.title('Alexnet-Cifar-10',fontsize=18)
plt.xlabel('epochs', fontsize=18)
plt.ylabel('mean of cosine similarity', fontsize=18)
#plt.show()

plt.subplot(1,3,2)
for layer in range(12):
    if (layer%2 == 0):
        plt.plot(epochs,np.var(data[:,:,layer],axis=0))
    
plt.legend([0,2,4,6,8,10], prop={'size': 18})
plt.title('Alexnet-Cifar-10',fontsize=18)
plt.xlabel('epochs', fontsize=18)
plt.ylabel('var of cosine similarity', fontsize=18)

plt.subplot(1,3,3)
for layer in range(12):
    if (layer%2 == 0):
        plt.plot(epochs,np.average(data[:,:,layer],axis=0)/np.var(data[:,:,layer],axis=0))
    
plt.legend([0,2,4,6,8,10], prop={'size': 18})
plt.title('Alexnet-Cifar-10',fontsize=18)
plt.xlabel('epochs', fontsize=18)
plt.ylabel('mean/var of cosine similarity', fontsize=18)

plt_name = 'cos_vis_alexnet_per_layer_nonmember.pdf'
plt.savefig(plt_name,dpi=800, bbox_inches="tight")
plt.show()
```
