```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import re
attack_features_name = ['random_guess','sum_grad_norm','sum_loss','sum_cos','sum_grad_diff',
                        'all_epoch_grad_norm','all_epoch_loss','all_epoch_cos','all_epoch_grad_diff',
                       'sum_4','all_epoch_4_features w/o loss redudction','sum_loss_reduction','all_epoch_all_5_features']
num_attacks = len(attack_features_name)
attack_name = ['all_user_mi','avg_user_mi','si','gmi']
#attack_name = ['avg_user_mi']
prefix = '/Users/jclialex/PycharmProjects/whiteboxmi_expdata/expdata/new_expdata'
matplotlib.rcParams.update({'font.size': 22})
color_list = ['b','g','r','c','m','y','k','indigo','brown','gray','purple','yellow','lime']

def attack_feature_vis(data_dict,attack_feature_name,attack_baseline=None,acc_index=None):
    
    num_attacks = len(attack_feature_name)
    num_cases = len(data_dict)
    
    fig = plt.figure(figsize=(20,15))
    
    for i,(key,val) in enumerate(data_dict.items()):
        attack_acc = np.array(val)[acc_index]
        plt.subplot(3,2,i+1)
    
        plt.plot(np.arange(num_attacks)*2-1,[attack_baseline for _ in range(len(np.arange(num_attacks)*2-1))])
        for idx,(this_attack_feature_name,this_attack_acc) in enumerate(zip(attack_feature_name,attack_acc)):
            plt.bar(idx*2,attack_acc[idx],width=0.8,color=color_list[idx])
    
        plt.xticks(ticks=np.arange(num_attacks)*2,labels=['' for _ in range(num_attacks)])
        plt.ylim(bottom=attack_baseline-5,top=np.amax(attack_acc)+5)
    
        plt.ylabel('attack_acc')
        
        if (i==0):
            plt.legend(attack_feature_name,bbox_to_anchor=(2.6,1.8),ncol=4)
            
        plt.title(key.replace('feature_vis',''))
        
    plt.show()
    
```

```python
def extract_number(this_str):
    #print (this_str)
    number = re.findall("\d+\.\d+", this_str)
    #print (number)
    
    if (len(number)!=1):
        print (this_str)
        return None
    
    return number[0]
        
def parse_result_file(file_name):
    all_user_acc = []
    avg_per_user_acc = []
    si_acc = []
    gmi_acc = []
    
    f = open(file_name,"r")
    all_user_flag = False
    
    while (True):
        this_line = f.readline()
        #print (this_line)
        #print ("Membership binary classification" in this_line)
        
        if (len(this_line) == 0):
            break
        
        if ("Membership binary classification" in this_line and (not all_user_flag)):
            #print (this_line)
            for i in range(num_attacks):
                this_line = f.readline()
                all_user_acc.append(float(extract_number(this_line)))
                #print (this_line)
            all_user_flag = True
            #print (all_user_acc)
            
        elif ("averaged binary membership inference" in this_line):
            for i in range(num_attacks):
                this_line = f.readline()
                avg_per_user_acc.append(float(extract_number(this_line)))
        
        elif ("member only case" in this_line and ':' not in this_line):
            for i in range(num_attacks):
                this_line = f.readline()
                si_acc.append(float(extract_number(this_line)))
        
        elif ("testing instances included" in this_line and ':' not in this_line):
            for i in range(num_attacks):
                this_line = f.readline()
                gmi_acc.append(float(extract_number(this_line)))
            break
            
    f.close()
    all_acc = [all_user_acc,avg_per_user_acc,si_acc,gmi_acc]
    print (all_acc)
    return all_acc
        
def all_results_vis(file_name):
    all_acc = parse_result_file(os.path.join(prefix,file_name))
    fig_name_prefix = file_name.replace('.txt','')
    
    for this_acc in all_acc:
        if (len(this_acc)!=num_attacks):
            return

    for index,this_attack_name in enumerate(attack_name):
        this_fig_name = fig_name_prefix + '_' + this_attack_name + '.pdf'
        attack_feature_vis(all_acc[index],attack_features_name,this_fig_name)
```
