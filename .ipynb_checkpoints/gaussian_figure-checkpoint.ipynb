```python
def std_test(size=1000):

    samples = [np.random.normal(size=size) for i in range(1000)]
    from scipy.spatial import distance

    cos = []
    for i in range(1000):
        for j in range(1000):
            if (i!=j):
                cos.append(1 - distance.cosine(samples[i],samples[j]))
            
    cos = np.array(cos)
    #print (np.average(cos),np.std(cos))
    return np.std(cos)

test_size = [100,300,500,1000,2000,5000,10000]
test_std = [std_test(x) for x in test_size]

print (test_size)
print (test_std)
```

```python
fig = plt.figure(figsize=(10,10))
plt.plot(1/np.sqrt(np.array(test_size)),test_std)
```

```python

member_gradient = np.load(prefix+'member_gradients.npy')

#member_gradient = member_gradient[:,-615936:-25700]
member_gradient = member_gradient[:,-65936:-25700]
print (member_gradient.shape)

member_gradient = np.nan_to_num(member_gradient)

this_sum = np.sum(member_gradient[0])
this_gradient = member_gradient[0]/this_sum

#print (this_gradient[:500])

fig = plt.figure(figsize=(10,5))
bins = np.linspace(np.amin(this_gradient),np.amax(this_gradient),20)
counts,_ = np.histogram(this_gradient,bins)
plt.plot(bins[1:],counts)
plt.title('PDF')
plt.show()

this_gradient = this_gradient[this_gradient!=0]
fig = plt.figure(figsize=(10,5))
bins = np.linspace(np.amin(this_gradient),np.amax(this_gradient),20)
counts,_ = np.histogram(this_gradient,bins)
plt.plot(bins[1:],counts)

plt.title('PDF')
plt.show()
```
