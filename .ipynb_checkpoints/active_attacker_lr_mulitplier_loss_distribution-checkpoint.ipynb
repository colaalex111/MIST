```python
import numpy as np
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import warnings
#warnings.filterwarnings('ignore') 
from sklearn.metrics import roc_auc_score
import torch.nn.functional as F
from scipy.special import softmax

plt.rcParams.update({'font.size': 14})
    
        
def loss_analysis(all_epoch_cosine,all_epoch_grad_diff,all_epoch_loss,all_epoch_target_loss,all_epoch_target_after_loss,label,stop_epoch,eval_data_size=50,barplot=False,zoom_in=False,zoom_in_ranges=None,track_loss=False,save_fig_path=''):
    eval_data_size = eval_data_size
    num_instance = all_epoch_cosine.shape[0]
    num_epoch = all_epoch_cosine.shape[1]
    num_user = all_epoch_cosine.shape[2]
    epochs = (np.arange(num_epoch)+1)*10
    member_index = np.arange(int(num_instance/2))
    nonmember_index = np.setdiff1d(np.arange(num_instance),member_index)
    members_loss = np.nan_to_num(all_epoch_target_after_loss[member_index,:,:])
    nonmembers_loss = np.nan_to_num(all_epoch_target_after_loss[nonmember_index,:,:])
        
    members_avg = np.average(np.squeeze(members_loss[:,:,0]),axis=0)
    members_var = np.var(np.squeeze(members_loss[:,:,0]),axis=0)
    nonmembers_avg = np.average(np.squeeze(nonmembers_loss[:,:,0]),axis=0)
    nonmembers_var = np.var(np.squeeze(nonmembers_loss[:,:,0]),axis=0)
    auc_list = []
    for epoch_idx,this_epoch in enumerate(epochs):
        this_epoch_member_loss = np.squeeze(members_loss[:,epoch_idx,0])
        this_epoch_nonmember_loss = np.squeeze(nonmembers_loss[:,epoch_idx,0])
            
        num_members = int(len(this_epoch_member_loss))
            
        label = np.concatenate((np.ones((num_members)),np.zeros((num_members))))
        loss = np.concatenate((this_epoch_member_loss,this_epoch_nonmember_loss))
        loss = np.nan_to_num(loss)
        #print (loss[:500])
        this_auc = roc_auc_score(label,-1*loss)
        auc_list.append(this_auc)
        
    auc_list = np.array(auc_list)
    
    print (f"Highest AUC score for a single epoch {max(np.max(auc_list),np.max(1-auc_list))}")
        
    fig = plt.figure(figsize=(15,5))
    #plt.plot(epochs,auc_list)
    plt.plot(epochs,members_avg)
    plt.plot(epochs,nonmembers_avg)
    #plt.yticks(np.arange(11)/10)
        #plt.errorbar(epochs,members_avg,yerr=members_var)
        #plt.errorbar(epochs,nonmembers_avg,yerr=nonmembers_var)
    plt.legend(['members','non-members'])
    plt.show()
    
    if (barplot):
        target_epoch = 25 if (num_epoch>10) else 5
        members_loss = np.squeeze(members_loss[:,target_epoch,0])
        #print (members_loss.shape)
        nonmembers_loss = np.squeeze(nonmembers_loss[:,target_epoch,0])
        
        member_avg = np.average(members_loss)
        member_var = np.var(members_loss)
        nonmember_avg = np.average(nonmembers_loss)
        nonmember_var = np.var(nonmembers_loss)
        #print (f"member avg {member_avg}, member var{member_var}, member std{np.sqrt(member_var)}")
        #print (f"nonmember avg {nonmember_avg},nonmember var{nonmember_var},nonmember std{np.sqrt(nonmember_var)}")
        #print (f"diff of mean {member_avg - nonmember_avg},member std{np.sqrt(member_var)},nonmember std{np.sqrt(nonmember_var)}")
        
        bins = np.linspace(0,4,100) # cifar
        if (num_epoch<10):
            bins = np.linspace(2.5,100,200) # texas
        
        members_counts,_ = np.histogram(members_loss,bins)
        nonmembers_counts,_ = np.histogram(nonmembers_loss,bins)
        width = (bins[1] - bins[0])/2
        #fig = plt.figure(figsize=(10,5))
        #plt.bar(bins[:-1],members_counts,width=width)
        #plt.bar(bins[:-1]+width,nonmembers_counts,width=width)
        #plt.legend(['members','nonmembers'])
        #plt.show()
        from scipy.stats import norm
        
        acc_members_counts = np.zeros_like(members_counts)
        acc_members_counts[0] = members_counts[0]
        for i in range(1,len(acc_members_counts)):
            acc_members_counts[i] = acc_members_counts[i-1] + members_counts[i]
        
        acc_nonmembers_counts = np.zeros_like(nonmembers_counts)
        acc_nonmembers_counts[0] = nonmembers_counts[0]
        for i in range(1,len(acc_members_counts)):
            acc_nonmembers_counts[i] = acc_nonmembers_counts[i-1] + nonmembers_counts[i]
        
        fig = plt.figure(figsize=(20,7*(1+len(zoom_in_ranges))))
        num_figures = 2 + 2*len(zoom_in_ranges)
        
        lower = np.amin(bins)
        upper = np.max(bins)
        
        #plt.subplot(int(num_figures/2),2,1)
        #plt.plot(bins[:-1],members_counts/num_instance*2)
        #plt.plot(bins[:-1],nonmembers_counts/num_instance*2)
        #plt.legend(['members','nonmembers'])
        #plt.ylabel('Frequency')
        #plt.xlabel('Loss')
        #plt.title(f'Loss Distribution, PDF, Loss Range:[{lower},{upper}]')
        #plt.show()
        
        plt.subplot(int(num_figures/2),2,2)
        m1 = member_avg
        std1 = np.sqrt(member_var)
        m2 = nonmember_avg
        std2 = np.sqrt(nonmember_var)
        plt.plot(bins[:-1],acc_members_counts/num_instance*2)
        plt.plot(bins[:-1],acc_nonmembers_counts/num_instance*2)
        plt.legend(['members','nonmembers'])
        plt.ylabel('Cumulative Frequency')
        plt.xlabel('Loss')
        plt.title(f'Loss Distribution, CDF, Loss Range:[{lower},{upper}]')
        
        #plt.show()
        
        print (f"maximum CDF difference {np.amax(np.abs(acc_members_counts - acc_nonmembers_counts))/num_instance*2}")
        
        
        if (zoom_in):
            
            for zoom_in_idx,(lower,upper) in enumerate(zoom_in_ranges):
            
                bins = np.linspace(lower,upper,300)
                bins = np.concatenate((np.array([0]),bins))
        
                members_counts,_ = np.histogram(members_loss,bins)
                nonmembers_counts,_ = np.histogram(nonmembers_loss,bins)
                width = (bins[2] - bins[1])/2
        
        
                acc_members_counts = np.zeros_like(members_counts)
                acc_members_counts[0] = members_counts[0]
                for i in range(1,len(acc_members_counts)):
                    acc_members_counts[i] = acc_members_counts[i-1] + members_counts[i]
        
                acc_nonmembers_counts = np.zeros_like(nonmembers_counts)
                acc_nonmembers_counts[0] = nonmembers_counts[0]
                for i in range(1,len(acc_members_counts)):
                    acc_nonmembers_counts[i] = acc_nonmembers_counts[i-1] + nonmembers_counts[i]
        
                #fig = plt.figure(figsize=(15,5))
        
                #plt.subplot(int(num_figures/2),2,3+zoom_in_idx*2)
                #plt.plot(bins[1:],members_counts/num_instance*2)
                #plt.plot(bins[1:],nonmembers_counts/num_instance*2)
                #plt.legend(['members','nonmembers'])
                #plt.ylabel('Frequency')
                #plt.xlabel('Loss')
                #plt.locator_params(axis='x', nbins=5)
                #plt.title(f'Loss Distribution, PDF, Loss Range:[{lower},{upper}]')
        
                plt.subplot(int(num_figures/2),2,4+zoom_in_idx*2)
                plt.plot(bins[1:],acc_members_counts/num_instance*2)
                plt.plot(bins[1:],acc_nonmembers_counts/num_instance*2)
                #plt.ylim(bottom=0,top=1)
                plt.legend(['members','nonmembers'])
                plt.ylabel('Cumulative Frequency')
                plt.xlabel('Loss')
                plt.locator_params(axis='x', nbins=5)
                plt.title(f'Loss Distribution, CDF, Loss Range:[{lower},{upper}]')
                print (f"maximum CDF difference {np.amax(np.abs(acc_members_counts - acc_nonmembers_counts))/num_instance*2}")
                
        if (save_fig_path):
            plt.savefig(save_fig_path,dpi=800,bbox_inches='tight')
        plt.show()
        

def calculate_distance(m1,std1,m2,std2):
  
    from scipy.stats import norm

    def solve(m1,m2,std1,std2):
      a = 1/(2*std1**2) - 1/(2*std2**2)
      b = m2/(std2**2) - m1/(std1**2)
      c = m1**2 /(2*std1**2) - m2**2 / (2*std2**2) - np.log(std2/std1)
      return np.roots([a,b,c])

    #Get point of intersect

    result = solve(m1,m2,std1,std2)    
    r = result[0]

    area = norm.cdf(r,m2,std2) + (1.-norm.cdf(r,m1,std1))
    return area


def transform_loss_to_prob(x):
    ### this can be rewritten as a map function
    ### we are using cross entropy loss for all tasks
    #new_x = np.power(2,-x)
    new_x = np.exp(-x)
    #print (np.average(new_x))
    if (np.average(new_x)>0.9999999):
        return np.zeros_like(x)
    else:
        return new_x

def all_analysis(epochs,name_prefix,common_mid_str,dataset,model,eval_data_size,special_layers=None,num_layers=12,num_user=5,target_data_size=2000,lr_mul=1,mislabel=False,stop_epoch=8,fixed_label=False,barplot=False,zoom_in=False,zoom_in_ranges=None,track_loss=False,save_fig_path='',a=0,b=0,c=0,adv_name=None):
    #target_data_size = 2000
    #eval_data_size = 50    
    #print (f"{lr_mul}")
    num_classes = 10
    num_step = target_data_size // 100
    
    total_data_num = (num_user)*eval_data_size*2
    all_epoch_cos = np.zeros((total_data_num,len(epochs),num_user))
    all_epoch_loss = np.zeros((total_data_num,len(epochs),num_user))
    all_epoch_target_loss = np.zeros((total_data_num,len(epochs),num_user))
    all_epoch_grad_diff = np.zeros((total_data_num,len(epochs),num_user))
    all_epoch_layer_cos = np.zeros((total_data_num,len(epochs),num_user,num_layers))
    all_epoch_layer_grad_diff = np.zeros((total_data_num,len(epochs),num_user,num_layers))
    all_epoch_target_after_loss = np.zeros((total_data_num,len(epochs),num_user))
    all_epoch_label = np.zeros(total_data_num)
    
    all_track_before_loss = np.zeros((total_data_num,len(epochs)))
    all_track_after_loss = np.zeros((total_data_num,len(epochs)))
    all_track_before_prediction = np.zeros((total_data_num,len(epochs),num_classes))
    all_track_after_prediction = np.zeros((total_data_num,len(epochs),num_classes))
    instance_label = np.zeros((total_data_num))
    
    for epoch_idx in range(len(epochs)):
        epoch = epochs[epoch_idx]
        mid_str = common_mid_str + \
                str(epoch+1)+'_'+str(dataset)+'_'+str(target_data_size) + \
                '_'+str(eval_data_size)+'_'+str(model)+'.npy' 
        
        data_name = prefix+'all_info_multi_party_member_attack_' + mid_str
        data = np.load(data_name,allow_pickle=True)
        label_name = prefix+'all_label_multi_party_member_attack_'+ mid_str
        label = np.load(label_name)
        loss_data_name = prefix+'loss_info_multi_party_member_attack_' + mid_str
        loss_data = np.load(loss_data_name,allow_pickle=True)
        loss_label_name = prefix+'loss_label_multi_party_member_attack_' + mid_str
        loss_label = np.load(loss_label_name,allow_pickle=True)
        
        ### data shape [num_user*2,eval_data_size,num_user,num_layers,num_features]
    
        target_loss_data_name = prefix+'target_model_before_loss_info_multi_party_member_attack_'+ mid_str
        target_loss_data = np.load(target_loss_data_name,allow_pickle=True)
        target_after_loss_data_name = prefix+'target_model_after_loss_info_multi_party_member_attack_' + mid_str
        target_after_loss_data = np.load(target_loss_data_name,allow_pickle=True)
        data = np.reshape(data,(-1,num_user,num_layers,7))
        label = np.reshape(label,(-1))
        loss_data = np.reshape(loss_data,(-1,num_user))
        loss_label = np.reshape(loss_label,(-1))
        target_loss_data = np.reshape(target_loss_data,(-1,1))
        target_loss_data = np.tile(target_loss_data,(1,num_user))
        target_after_loss_data = np.reshape(target_after_loss_data,(-1,1))
        target_after_loss_data = np.tile(target_after_loss_data,(1,num_user))
        
        all_epoch_layer_cos[:,epoch_idx,:,:] = data[:,:,:,0]
        all_epoch_layer_grad_diff[:,epoch_idx,:,:] = data[:,:,:,1] - data[:,:,:,2]
        all_epoch_loss[:,epoch_idx,:] = loss_data
        all_epoch_target_loss[:,epoch_idx,:] = target_loss_data
        all_epoch_target_after_loss[:,epoch_idx,:] = target_after_loss_data
        all_epoch_label = label
        
        if (track_loss):
            loss_before_name = prefix+'active_track_loss_info_multi_party_member_attack_before_' + mid_str
            loss_before = np.load(loss_before_name)
            loss_after_name = prefix+'active_track_loss_info_multi_party_member_attack_after_' + mid_str
            loss_after = np.load(loss_after_name)
            before_prediction_name = prefix+'active_track_prediction_info_multi_party_member_attack_before_' + mid_str
            before_prediction = np.load(before_prediction_name)
            after_prediction_name = prefix+'active_track_prediction_info_multi_party_member_attack_after_' + mid_str
            after_prediction = np.load(after_prediction_name)
            
            all_track_before_loss[:,epoch_idx] = loss_before.flatten()
            all_track_after_loss[:,epoch_idx] = loss_after.flatten()
            all_track_before_prediction[:,epoch_idx,:] = np.reshape(before_prediction,(-1,num_classes))
            all_track_after_prediction[:,epoch_idx,:] = np.reshape(after_prediction,(-1,num_classes))
            
            instance_label_name = prefix+'acive_track_loss_label_multi_party_member_attack_after_'+ mid_str
            instance_label = np.load(instance_label_name)
        
        
    eval_data_size = eval_data_size
    num_instance = all_epoch_cos.shape[0]
    num_epoch = all_epoch_cos.shape[1]
    num_user = all_epoch_cos.shape[2]
    epochs = (np.arange(num_epoch)+1)*10
    member_index = np.arange(int(num_instance/2))
    nonmember_index = np.setdiff1d(np.arange(num_instance),member_index)
    
    members_loss = np.nan_to_num(all_epoch_target_after_loss[member_index,:,:])
    nonmembers_loss = np.nan_to_num(all_epoch_target_after_loss[nonmember_index,:,:])    
    #members_loss = np.nan_to_num(all_epoch_cos[member_index,:,:])
    #nonmembers_loss = np.nan_to_num(all_epoch_cos[nonmember_index,:,:])
        
    members_avg = np.average(np.squeeze(members_loss[:,:,0]),axis=0)
    members_var = np.var(np.squeeze(members_loss[:,:,0]),axis=0)
    nonmembers_avg = np.average(np.squeeze(nonmembers_loss[:,:,0]),axis=0)
    nonmembers_var = np.var(np.squeeze(nonmembers_loss[:,:,0]),axis=0)
    
    
    if (barplot):
        target_epoch = 25 if (num_epoch>10) else 3
        members_loss = np.squeeze(members_loss[:,target_epoch,0])
        nonmembers_loss = np.squeeze(nonmembers_loss[:,target_epoch,0])
        
        member_avg = np.average(members_loss)
        member_var = np.var(members_loss)
        nonmember_avg = np.average(nonmembers_loss)
        nonmember_var = np.var(nonmembers_loss)
        
        print (f"member avg {member_avg}, nonmember avg {nonmember_avg}")
        
        bins = np.linspace(-0.5,10,100) # cifar
        if (num_epoch<10):
            bins = np.linspace(2.5,3,50) # texas
        
        members_counts,_ = np.histogram(members_loss,bins)
        nonmembers_counts,_ = np.histogram(nonmembers_loss,bins)
        width = (bins[1] - bins[0])/2
        
        acc_members_counts = np.zeros_like(members_counts)
        acc_members_counts[0] = members_counts[0]
        for i in range(1,len(acc_members_counts)):
            acc_members_counts[i] = acc_members_counts[i-1] + members_counts[i]
        
        acc_nonmembers_counts = np.zeros_like(nonmembers_counts)
        acc_nonmembers_counts[0] = nonmembers_counts[0]
        for i in range(1,len(acc_members_counts)):
            acc_nonmembers_counts[i] = acc_nonmembers_counts[i-1] + nonmembers_counts[i]

        lower = np.amin(bins)
        upper = np.max(bins)
        
        #plt.subplot(a,b,c+1)
        #plt.plot(bins[:-1],members_counts/num_instance*2)
        #plt.plot(bins[:-1],nonmembers_counts/num_instance*2)
        #plt.legend(['members','nonmembers'])
        #plt.ylabel('Frequency')
        #plt.xlabel('Loss')
        #plt.title(f'({chr(ord("a")+c)}) Loss Distribution - PDF')
        #plt.show()
        plt.subplot(a,b,c+1)
        m1 = member_avg
        std1 = np.sqrt(member_var)
        m2 = nonmember_avg
        std2 = np.sqrt(nonmember_var)
        plt.plot(bins[1:],acc_members_counts/num_instance*2)
        plt.plot(bins[1:],acc_nonmembers_counts/num_instance*2)
        plt.plot(bins[1:],np.abs(acc_members_counts/num_instance*2 - acc_nonmembers_counts/num_instance*2))
        #plt.legend(['members','nonmembers','difference'])
        plt.ylabel('Cumulative Frequency')
        plt.xlabel('Loss')
        fig_idx = chr(ord('a')+c)
        plt.title(f'({fig_idx}) {adv_name}, loss range (0,10)')
        plt.ylim(-0.03,1.03)
        
        print (f"maximum CDF difference {np.amax(np.abs(acc_members_counts - acc_nonmembers_counts))/num_instance*2}")
        #print (acc_members_counts - acc_nonmembers_counts)
        
        if (zoom_in_ranges is not None):
            ## transform member loss to normal distributed.
            #members_loss = np.log(np.exp(-1*members_loss)/(1-np.exp(-1*members_loss)+1e-8))
            #nonmembers_loss = np.log(np.exp(-1*nonmembers_loss)/(1-np.exp(-1*nonmembers_loss)+1e-8))
            #print (np.amax(members_loss),np.amin(members_loss))
            #print (np.amax(nonmembers_loss),np.amin(nonmembers_loss))
            
            for zoom_in_idx,(lower,upper) in enumerate(zoom_in_ranges):
                
                #print (lower,upper)
                bins = np.linspace(lower,upper,60)
                #bins = np.concatenate((np.array([0]),bins))
        
                members_counts,_ = np.histogram(members_loss,bins)
                nonmembers_counts,_ = np.histogram(nonmembers_loss,bins)
                width = (bins[2] - bins[1])/2
        
                acc_members_counts = np.zeros_like(members_counts)
                acc_members_counts[0] = members_counts[0]
                for i in range(1,len(acc_members_counts)):
                    acc_members_counts[i] = acc_members_counts[i-1] + members_counts[i]
        
                acc_nonmembers_counts = np.zeros_like(nonmembers_counts)
                acc_nonmembers_counts[0] = nonmembers_counts[0]
                for i in range(1,len(acc_members_counts)):
                    acc_nonmembers_counts[i] = acc_nonmembers_counts[i-1] + nonmembers_counts[i]
        
                #plt.subplot(a,b,c+3)
                #plt.plot(bins[1:],members_counts/num_instance*2)
                #plt.plot(bins[1:],nonmembers_counts/num_instance*2)
                #plt.legend(['members','nonmembers'])
                #plt.ylabel('Frequency')
                #plt.xlabel('Loss')
                #plt.locator_params(axis='x', nbins=5)
                #plt.title(f'({chr(ord("a")+c+2)}) Loss Distribution - PDF')
                
                plt.subplot(a,b,c+2)
                plt.plot(bins[1:],acc_members_counts/num_instance*2)
                plt.plot(bins[1:],acc_nonmembers_counts/num_instance*2)
                plt.plot(bins[1:],np.abs(acc_members_counts/num_instance*2 - acc_nonmembers_counts/num_instance*2))
                #plt.ylim(bottom=0,top=1)
                #plt.legend(['members','nonmembers','difference'])
                plt.ylabel('Cumulative Frequency')
                plt.xlabel('Loss')
                plt.locator_params(axis='x', nbins=5)
                fig_idx = chr(ord('a')+c+1)
                plt.title(f'({fig_idx}) {adv_name}, loss range (0,0.0003)')
                print (f"maximum CDF difference {np.amax(np.abs(acc_members_counts - acc_nonmembers_counts))/num_instance*2}")
                plt.ylim(-0.03,1.03)
                #print (acc_members_counts - acc_nonmembers_counts)
        #if (save_fig_path):
        #    plt.savefig(save_fig_path,dpi=800,bbox_inches='tight')
        
        
        
        
```
