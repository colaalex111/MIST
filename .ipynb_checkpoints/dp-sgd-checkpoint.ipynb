```python
import os
import copy
import numpy as np
arr = os.listdir('./expdata/')
#print(arr)
batch_attack_types=['cosine_batch','grad_minus_batch_l1diff','grad_in_out_batch_l1diff']
batch_non_member_attack_types = ['cosine_batch_non_member',
                                 'grad_minus_batch_non_member_l1diff',
                                 'grad_in_out_batch_non_member_l1diff']
epoch_attack_types = ['cosine_epoch','grad_minus_epoch_l1diff','grad_in_out_epoch_l1diff']
epoch_non_member_attack_types =['cosine_epoch_non_member','grad_minus_peoch_non_member_l1diff','grad_in_out_epoch_non_member_l1diff']
epochs = [40,80,120,160]
batch_size = 100

for this_attack_type in batch_non_member_attack_types:
    for epoch in epochs:
        print(this_attack_type,epoch)
        #dpsgd_data_file = []
        dpsgd_param = []
        dpsgd_auc = []
        #mmd_data_file = []
        mmd_param = []
        mmd_auc = []
        for data_file in arr:
            
            data_file_name = copy.deepcopy(data_file)
            
            if (this_attack_type in data_file and str(epoch) in data_file):
                #print (data_file)
                
                this_file_name = data_file.replace(this_attack_type+'_','')
                #print(this_file_name)
                this_file_name = this_file_name.replace(str(epoch)+'_cifar100_5000.npy','')
                #print (str(epoch)+'_cifar100_5000.npy')
                #print (this_file_name)
                
                ### get the dpsgd param 
                #print (this_file_name.find('e'))
                if (this_file_name.find('e')==-1):
                    continue
                noise_scale = this_file_name[0:this_file_name.find('e')+4]
                dpsgd_sign=noise_scale[0]
                noise_scale = noise_scale[1:]
                this_file_name=this_file_name[this_file_name.find('e')+4:]
                #print (this_file_name)
                ### get the mmd param
                mmd_sign=this_file_name[0]
                this_file_name=this_file_name[1:]
                mixup_sign=this_file_name[-1]
                mmd_scale=this_file_name[:-1]
                
                #print (dpsgd_sign,noise_scale,mmd_sign,mmd_scale,mixup_sign)
                
                if (dpsgd_sign == '1'):
                    
                    if (noise_scale[0]=='_'):
                        #noise_scale = noise_scale[1:]
                        continue
                    #else:
                    #    continue
                    if (float(noise_scale)>=1e-10):
                        continue
                    
                    dpsgd_param.append(float(noise_scale))
                    this_file_data = np.load('./expdata/'+data_file_name)
                    #print (this_file_data.shape)
                    this_file_data = np.reshape(this_file_data,(-1,this_file_data.shape[-1]))
                    layer_num=this_file_data.shape[-1]
                    #dpsgd_data_file.append(this_file_data)
                    ## compute auc score
                    
                    from sklearn.metrics import roc_auc_score
                    data_size = this_file_data.shape[0]/2
                    y_true = []
                    if ('batch' in this_attack_type):
                          for j in range(int(data_size/batch_size)):
                            y_true.append(np.concatenate((np.ones(batch_size),np.zeros(batch_size))))
                    else:
                        y_true = np.concatenate((np.ones(data_size),np.zeros(data_size)))
                    y_true = np.array(y_true).flatten()
                    max_auc = 0
                    for layer_idx in range(layer_num):
                        y_pred = this_file_data[:,layer_idx]
                        max_auc = max(max_auc,roc_auc_score(y_true,y_pred))
                        
                        #print (np.log10(float(noise_scale)))
                        #if (np.log10(float(noise_scale))>=(-10)):
                        #    print (np.log10(float(noise_scale)),layer_idx,max_auc)
                    dpsgd_auc.append(max_auc)
                    
                    
                    #print (noise_scale,max_auc)
                
                if (mmd_sign == '1'):
                    mmd_param.append(float(mmd_scale))
                    this_file_data = np.load('./expdata/'+data_file_name)
                    #print (this_file_data.shape)
                    this_file_data = np.reshape(this_file_data,(-1,this_file_data.shape[-1]))
                    layer_num=this_file_data.shape[-1]
                    #mmd_data_file.append(this_file_data)
                    ## compute auc score
                    
                    from sklearn.metrics import roc_auc_score
                    data_size = this_file_data.shape[0]/2
                    y_true = []
                    if ('batch' in this_attack_type):
                          for j in range(int(data_size/batch_size)):
                            y_true.append(np.concatenate((np.ones(batch_size),np.zeros(batch_size))))
                    else:
                        y_true = np.concatenate((np.ones(data_size),np.zeros(data_size)))
                    y_true = np.array(y_true).flatten()
                    max_auc = 0
                    for layer_idx in range(layer_num):
                        y_pred = this_file_data[:,layer_idx]
                        max_auc = max(max_auc,roc_auc_score(y_true,y_pred))
                    mmd_auc.append(max_auc)
                    
                    #print (mmd_scale,max_auc)
                    
                    
            ### after getting all the auc score and the param
            
        import matplotlib.pyplot as plt
        fig = plt.figure(figsize=(4,4))
        
        #plt.subplot(1,2,1)
        #print (len(mmd_param),len(mmd_auc),len(dpsgd_param),len(dpsgd_auc))
        #plt.scatter(np.array(mmd_param),np.array(mmd_auc))  
        #plt.subplot(1,2,2)
        plt.scatter(np.log10(np.array(dpsgd_param)),np.array(dpsgd_auc))
        plt.show()
        #sorted_index = np.argsort(np.array(dpsgd_param))
        #dpsgd_param=np.array(dpsgd_param)[sorted_index]
        #dpsgd_auc=np.array(dpsgd_auc)[sorted_index]
        #for i in range(len(dpsgd_param)):
        #    print (dpsgd_param[i],dpsgd_auc[i])
            
        
```

```python
epochs = [40,80]
import numpy as np
import matplotlib.pyplot as plt
data_size = 5000
batch_size = 100
user_num = 6
for epoch in epochs:
    
    
    data_name = './expdata/grad_minus_batch_non_member_allinfo_1_1e-13_0_0.0_0_'+str(epoch)+'_cifar100_'+str(data_size)+'.npy'
    data = np.load(data_name)
    print (data.shape)

    label = []
    for idx in range(int(data_size*6/(batch_size*2))):
        label.append(np.ones(batch_size))
        label.append(np.zeros(batch_size))
    label = np.array(label).flatten()
    member_index = np.arange(data_size*user_num)[label == 1]
    nonmember_index = np.arange(data_size*user_num)[label == 0]
    member_data = data[member_index,:,:]
    nonmember_data = data[nonmember_index,:,:]
    
    layers = [-2]
    for layer in layers:
        
        
        nonmember_small_norm_index = []
        
        this_layer_member_data = member_data[:,layer,:]
        this_layer_nonmember_data = nonmember_data[:,layer,:]
        
        import matplotlib.pyplot as plt
        
        fig = plt.figure(figsize=(20,5))
        
        num_bins = 300
        
        for data_type in range(3):
            #print (epoch,layer,data_type)
            plt.subplot(1,4,data_type+1)
            this_type_member_data = this_layer_member_data[:,data_type]
            this_type_nonmember_data = this_layer_nonmember_data[:,data_type]
            #print (this_type_member_data.shape)
    
            counts,bins=np.histogram(this_type_member_data,num_bins)
            width = bins[1] - bins[0]
            plt.bar(bins[1:],counts,color='red',width=width)
            counts,bins=np.histogram(this_type_nonmember_data,bins)
            plt.bar(bins[1:],counts,color='blue',width=width)
            
            plt.legend(['member','non-member'])
            from sklearn.metrics import roc_auc_score
            y_pred=np.concatenate((this_type_member_data,this_type_nonmember_data))
            y_true=np.concatenate((np.ones(len(this_type_member_data)),np.zeros(len(this_type_nonmember_data))))
            auc_score=roc_auc_score(y_true,y_pred)
        
            print (auc_score)    
            #print (np.median(this_type_member_data),np.median(this_type_nonmember_data))
            
            if(data_type == 2):
                for i in range(len(this_type_nonmember_data)):
                    if (this_type_nonmember_data[i]<np.median(this_type_member_data)):
                        nonmember_small_norm_index.append(i)
            
            
            
    
        plt.subplot(1,4,4)
        this_type_member_data = this_layer_member_data[:,0] - this_layer_member_data[:,1]
        this_type_nonmember_data = this_layer_nonmember_data[:,0] - this_layer_nonmember_data[:,1]
    
        counts,bins=np.histogram(this_type_member_data,num_bins)
        width = bins[1] - bins[0]
        plt.bar(bins[1:],counts,color='red',width=width)
        counts,bins=np.histogram(this_type_nonmember_data,bins)
        plt.bar(bins[1:],counts,color='blue',width=width)
        
        ### also drawing some nonmember smallnorm results
        print (len(nonmember_small_norm_index))
        counts,bins=np.histogram(this_type_nonmember_data[nonmember_small_norm_index],bins)
        plt.bar(bins[1:],counts,color='green',width=width)
        
        
            
        from sklearn.metrics import roc_auc_score
        y_pred=np.concatenate((this_type_member_data,this_type_nonmember_data))
        y_true=np.concatenate((np.ones(len(this_type_member_data)),np.zeros(len(this_type_nonmember_data))))
        auc_score=roc_auc_score(y_true,y_pred)
        print (auc_score)
        print (np.sum(this_type_member_data)/15000,np.sum(this_type_nonmember_data)/15000)
        plt.legend(['member','non-member','non-member-small-grad-norm'])
        
        
        
        plt.show()
    

```
