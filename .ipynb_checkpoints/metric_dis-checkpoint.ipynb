```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.metrics import roc_curve

def show_dis(data_name,label_name):
    metric =np.load(data_name)
    label = np.load(label_name)
    print (np.amax(metric),np.amin(metric))
    #rand_index = np.random.choice(len(metric),100000,replace=False)
    rand_index = np.arange(len(label))
    metric = metric[rand_index]
    label = label[rand_index]
    print (len(metric),len(label))
    fpr, tpr, thresholds = roc_curve(label, metric, pos_label=1)
   
    fpr = fpr[::-1]
    tpr = tpr[::-1]
    d = {}
    for x,y in zip(fpr,tpr):
        if (x in d):
            continue
        d[x] = y
    new_fpr = list(d.keys())
    new_tpr = list(d.values())
    fpr = np.array(new_fpr)[::-1]
    tpr = np.array(new_tpr)[::-1]
    #if ('200' in data_name):
    #    print (fpr[:400])
    #    print (tpr[:400])
    ## remove fpr < 1e-5
    #for c in range(len(fpr)):
    #    if (fpr[c]>=1e-5):
    #        break
    #fpr = fpr[c:]
    #tpr = tpr[c:]
    
    uniform_fpr = [1e-4] + [1e-4*i*2 for i in range(1,6)] + [1e-3*i/2 for i in range(2,20)]
    uniform_fpr = uniform_fpr + [1e-2*i/2 for i in range(2,200)] #+ [1e-1*i for i in range(1,11)]
    uniform_tpr = []
    for _,this_fpr in enumerate(uniform_fpr):
        for j in range(len(fpr)):
            if (fpr[j]>=this_fpr):
                this_tpr = tpr[j]
                break
        uniform_tpr.append(this_tpr)
        
    uniform_fpr = np.array(uniform_fpr)
    uniform_tpr = np.array(uniform_tpr)
    
    
    plt.subplot(1,2,1)
    plt.plot(uniform_fpr,uniform_tpr)
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title('ROC curve')
    plt.legend(['w/ cross-diff-loss', 'w/o cross-diff-loss'])
    plt.subplot(1,2,2)
    plt.plot(np.log10(uniform_fpr),(uniform_tpr/uniform_fpr))
    plt.xlabel('FPR')
    plt.ylabel('PLR')
    plt.title('PLR v.s. FPR')
    #plt.show()
```

```python
def show_fpr_dis(data_name,label_name,fpr=1e-5):
   
    metric =np.load(data_name)
    label = np.load(label_name)
    print (len(metric))
    member_index = np.arange(len(label))[label == 1]
    nonmember_index = np.arange(len(label))[label == 0]
    #print (len(member_index))
    threshold = np.sort(metric[nonmember_index])[::-1][int(fpr*len(nonmember_index))]
    #print (threshold)
    left_index = np.arange(len(label))[metric > threshold]

    metric = metric[left_index]
    label = label[left_index]
    metric = np.log10(metric)
    bins = np.linspace(np.amin(metric),np.amax(metric),num=10)
    member_index = np.arange(len(label))[label == 1]
    nonmember_index = np.arange(len(label))[label == 0]
    member_metric = metric[member_index]
    nonmember_metric = metric[nonmember_index]
    member_count,_ = np.histogram(member_metric,bins)
    nonmember_count,_ = np.histogram(nonmember_metric,bins)

    fig = plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.plot(bins[1:],member_count)
    plt.plot(bins[1:],nonmember_count)
    plt.legend(['member','nonmember'])
    plt.subplot(1,2,2)

    plt.plot(bins[1:],member_count/nonmember_count)
    plt.show()
    print (threshold,np.sum(member_count),np.sum(nonmember_count))
```
